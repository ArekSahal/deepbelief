```
import sys
path = "/content/drive/MyDrive/bm/"
sys.path.append(path)
```

```
import numpy as np
import torch
import matplotlib.pyplot as plt
from dbimage_torch import DBImage
plt.rcParams['image.cmap'] = 'Greys_r'
print("Loading MNIST dataset")
with open(path+'data/train_data.pt', 'rb') as f:
    train_X = torch.load(f)
with open(path+'data/train_labels.pt', 'rb') as f:
    train_labels = torch.load(f)
with open(path+'data/test_data.pt', 'rb') as f:
    test_X = torch.load(f)
with open(path+'data/test_labels.pt', 'rb') as f:
    test_labels = torch.load(f)


print("Done loading MNIST")
n_train = 60000
n_test = 10000

test_X = test_X[:n_test,:,:].float()
train_X = train_X[:n_train,:,:].float()
test_labels = test_labels[:n_test].float()
train_labels = train_labels[:n_train].float()

cdns = [1,1,1]
```

```
model = DBImage(28,28,10,[500,500,2000],
                batch_size=20,
                learning_rate=0.005,
                momentum=0.9,
                initial_momentum=0.5,
                weight_decay=2*10**(-4),
                tune_momentum=0.7,
                tune_rate=0.001, 
                tune_decay=0.000001, 
                cdns=cdns,
                sparsity_target=0.05,
                sparsity_penalty=0.0,
                path_writer=path+"runs/MNIST/Final/pre_train_long")

model.set_images(train_X, train_labels,split=0.98)

#model.load_weights(path+"models/db_")


model.train(epochs=50,save=True, filename=path + "models/db_", full_loss=True)
model.validate(test_X,test_labels, full=True, no_print=False, generate=True)
```

```

model = DBImage(28,28,10,[500,500,2000],
                batch_size=20,
                learning_rate=0.01,
                momentum=0.9,
                initial_momentum=0.5,
                weight_decay=2*10**(-4),
                tune_momentum=0.7,
                tune_rate=0.0005, 
                tune_decay=0.00001, 
                sparsity_target=0.05,
                sparsity_penalty=0.0,
                path_writer=path+"runs/MNIST/Final/tuned_long")

model.set_images(train_X, train_labels,split=0.98)

model.load_weights(path+"models/db_")

model.tune([100,100,100],save=True,filename=path+"models/db_tuned_",cd=[3,6,10])
model.validate(test_X,test_labels,full=True, generate=True)
```
